```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "diabetes.csv"  # Change this to the path where you saved the downloaded file
column_names = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"]
diabetes_df = pd.read_csv(file_path, names=column_names)
```

```python
print("First few rows of the dataset:")
print(diabetes_df.head())
```

```python
print("\nLast few rows of the dataset:")
print(diabetes_df.tail())
```

```python
print("\nData types of columns:")
print(diabetes_df.dtypes)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Glucose', hue='Outcome', data=diabetes_df)
plt.title('Glucose Level vs. Age for Diabetic and Non-Diabetic Patients')
plt.xlabel('Age')
plt.ylabel('Glucose Level')
plt.show()
```
